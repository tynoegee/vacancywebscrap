```python
import scrapy

class JobsSpider(scrapy.Spider):
    name = "jobs"
    start_urls = [
        'https://hotzimbabwejobs.com/',
    ]
    custom_settings = {
        'FEED_FORMAT': 'csv',
        'FEED_URI': 'jobs.csv'
    }

    def parse(self, response):
        for job in response.css('div.job-info'):
            yield {
                'job title': job.css('span.job-title ::text').get(),
                'Company Name': job.css('span.company-name ::text').get(),
                'Job Link': job.css('h4 a::attr(href)').extract()[0],
                
            }

        next_page = response.css('li.list-item a.next page-numbers a::attr(href)').get()
        if next_page is not None:
            yield response.follow(next_page, self.parse)
```

```python
from scrapy.crawler import CrawlerProcess

process1 = CrawlerProcess(settings={
    'FEED_FORMAT': 'csv',
    'FEED_URI': 'jobs.csv'
})
process1.crawl(JobsSpider)
process1.start()
process1.stop()
```
